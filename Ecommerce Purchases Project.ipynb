## Lets make a chart

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6, 4))
order_counts = data['AM or PM'].value_counts()
order_counts.plot(kind='bar', color=['blue', 'orange'])

# Labels and title
plt.xlabel('Time of Day')
plt.ylabel('Number of Purchases')
plt.title('Number of Purchases in AM vs PM')
plt.xticks(rotation=0)  

# Show plot
plt.show()
```

# Top 5 most popular providers (e.g gmail.com,yahoo.com)

```python
list1 = []

for email in data['Email']:
    list1.append(email.split('@')[1])
```

```python
data['temp'] = list1
```

```python
data.head(1)
```

```python
data['temp'].value_counts().head()
```

```python


top_providers = data['temp'].value_counts().head()

plt.figure(figsize=(8, 5))
top_providers.plot(kind='bar', color=['blue', 'orange', 'green', 'red', 'purple'])

# Labels and title
plt.xlabel('Email Provider')
plt.ylabel('Number of Purchases')
plt.title('Top 5 Email Providers Used for Purchases')
plt.xticks(rotation=30)  # Rotate labels for better visibility

# Show plot
plt.show()
```
